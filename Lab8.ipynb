```python
Контрольні запитання
Ви хочете заборонити користувачеві модуля запускати ваш код як звичайний скрипт. Як досягти такого ефекту? Написати код.
Щоб заборонити користувачеві запускати код у модулі як скрипт, потрібно використати перевірку змінної __name__. Якщо модуль імпортується, значення __name__ буде рівним імені модуля. Якщо код запускається як скрипт, значення буде "__main__"
if __name__ == "__main__":
    raise RuntimeError("Цей модуль не можна запускати як скрипт.")

Деякі додаткові та необхідні пакети зберігаються в каталозі D:. Напишіть код, який забезпечує перегляд каталогу Python для пошуку всіх модулів. Написати код.

Щоб переглянути вміст каталогу, можна скористатися модулем os. Код нижче шукає всі файли Python у каталозі D: і виводить їх список.
import os

directory = "D:"
python_files = []

for root, dirs, files in os.walk(directory):
    for file in files:
        if file.endswith(".py"):
            python_files.append(os.path.join(root, file))

print("Знайдені модулі Python:")
for f in python_files:
    print(f)


```
