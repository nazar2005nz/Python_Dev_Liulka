Написати програму, яка виводила б на екран наступну картинку:


```python
Вигляд коду
def draw_arrow():
    arrow = [
        "   *   ",
        "  * *  ",
        " *   * ",
        "***  ***",
          "*  *",
          "*  *",
          "*****"
    ]
    for line in arrow:
        print(line)

draw_arrow()
```
